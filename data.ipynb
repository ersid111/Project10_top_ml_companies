```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
from autoscraper import AutoScraper
```

```python
Name=[]
info=[]
location=[]
Website=[]
len_loc=[]
for i in range(40):
    url = "https://clutch.co/developers/artificial-intelligence/machine-learning?page="+str(i)
    page = requests.get(url, verify=False)
    page_text =page.text

    Comany_name = (re.findall('data-title="(.+?)"', page_text))
    company_info = (re.findall('class="company_info__wrap tagline">(.+?)</p>', page_text))
    company_location=(re.findall('<span class="locality">(.+?)</span>', page_text))
    website=(re.findall('href=http*(.+?)>',page_text))


    Name=Name+Comany_name
    info=info+company_info
    location=location+company_location
    Website=Website+website
    len_loc.append(len(company_location))
```
